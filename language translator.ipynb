```python
from googletrans import Translator, LANGUAGES

def print_language_names():
    print("Supported target languages (use these names):\n")
    for lang in sorted(LANGUAGES.values()):
        print(f"- {lang}")
    print("\nEnter the target language by name (e.g., english, french, spanish).")

def get_language_code(lang_name):
    for code, name in LANGUAGES.items():
        if name.lower() == lang_name.lower():
            return code
    return None

def translate_text(text, dest_lang_name):
    dest_code = get_language_code(dest_lang_name)
    if not dest_code:
        return None, None, f"Target language '{dest_lang_name}' not supported."
    
    translator = Translator()
    try:
        result = translator.translate(text, dest=dest_code)
        detected_lang_code = result.src
        detected_lang_name = LANGUAGES.get(detected_lang_code, "Unknown")
        return result.text, detected_lang_name, None
    except Exception as e:
        return None, None, f"Error: {str(e)}"

# Main flow
print_language_names()
text_to_translate = input("Enter the text to translate: ")
target_language_name = input("Enter target language name (default is 'english'): ") or 'english'

translated_text, detected_lang, error = translate_text(text_to_translate, target_language_name)

if error:
    print(error)
else:
    print(f"\n‚úÖ Detected Source Language: {detected_lang}")
    print(f"üåç Translated Text: {translated_text}")
```
